```python
# Install required library
!pip install transformers

# Import libraries
import pandas as pd
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import random
import torch

# Check for GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load GPT-2 model and tokenizer
model_name = "gpt2"  # Use "distilgpt2" for faster computation
tokenizer = GPT2Tokenizer.from_pretrained(model_name)
model = GPT2LMHeadModel.from_pretrained(model_name).to(device)
model.eval()

# Function to calculate perplexity for a sequence
def calculate_perplexity(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, max_length=1000).to(device)
    with torch.no_grad():
        outputs = model(**inputs, labels=inputs["input_ids"])
    loss = outputs.loss.item()
    perplexity = torch.exp(torch.tensor(loss))
    return perplexity.item()

# Beam Search function with random sampling
def random_beam_search(text, beam_width=10, max_candidates=20):
    words = text.split()
    best_sequence = text
    best_score = calculate_perplexity(text)

    for _ in range(max_candidates):
        # Shuffle words randomly
        random.shuffle(words)
        candidate_text = " ".join(words)
        perplexity = calculate_perplexity(candidate_text)

        # Update best sequence
        if perplexity < best_score:
            best_sequence = candidate_text
            best_score = perplexity

    return best_sequence

# Load sample_submission.csv
sample = pd.read_csv("/kaggle/input/santa-2024/sample_submission.csv")

# Process each text entry
def process_texts(df, beam_width=3, max_candidates=10):
    processed_texts = []
    for idx, row in df.iterrows():
        text = row['text']
        optimized_text = random_beam_search(text, beam_width=beam_width, max_candidates=max_candidates)
        processed_texts.append(optimized_text)
        if (idx + 1) % 10 == 0:  # Print progress every 10 rows
            print(f"Processed {idx + 1}/{len(df)} texts")
    return processed_texts

# Apply beam search to the sample data
sample['text'] = process_texts(sample, beam_width=3, max_candidates=5)

# Save the submission file
sample.to_csv("submission.csv", index=False)
print("Submission saved as submission.csv")
```
